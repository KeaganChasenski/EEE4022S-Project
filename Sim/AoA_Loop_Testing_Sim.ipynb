```julia
using FFTW

function l(i)
    f0 = 150e6  # 150 MHz
    fs = 350e8 # Sampling 300MHz
    Δt = 1/fs
    t = 0:Δt:3/f0
    ω0 = 2*pi*f0;   # rad/s   
    A = 1

    v1 = A*sin.(ω0*t);
    v2 = A*sin.(ω0*t.-i);

    V1 = fft(v1)
    V2 = fft(v2)

    maxsignal_1 = findmax(abs.(V1))
    maxsignal_2 = findmax(abs.(V2))

    maxsignal_1 = maxsignal_1[2]
    maxsignal_2 = maxsignal_2[2]

    p1 = V1[maxsignal_1]
    p2 = V2[maxsignal_2]

    println("here",imag(p1))
    println((p2))
    
    ### Using conjugate
    Z = p1 .* conj(p2)
    phase_diff = angle(Z) # in rads

    y = 1.998616 # Wave length and 89MHz
    d = 1 # 119cm 1.19m spacing of antenna


    println("pd",phase_diff) # in rads
    x = (y * phase_diff) / (2 * pi * d)

    AoA_rad = asin(x)
    #println("Angle of arrival in rad: ", AoA_rad)
    AoA_deg = (AoA_rad * 180/pi)

    println(AoA_deg)
    #println("Phase shift rads = ", i)
    println((90 - (i * 180/pi)))
    println("#################### \n")
end
```

```julia
i = 0
while i <= (pi/2)
    l(i)
    global i += 0.0872665
end
```
